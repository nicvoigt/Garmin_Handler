```python
from Garmin_Handler.Handler import Overview_File_Reader, Activity_Handler
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
activities = Overview_File_Reader().read_file()
run_data = Activity_Handler(activities).run_data()
```

```python
test_cluster = run_data.loc[:,("Kalorien", "Maximale Herzfrequenz")]
test_cluster.dropna(axis = 0, inplace = True)
test_cluster.reset_index(drop = True, inplace = True)
test_cluster
```

```python
X = np.array(test_cluster)
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
kmeans.labels_
```

```python
tc_labels = pd.concat([test_cluster, pd.Series(kmeans.labels_, name = "label")], axis = 1)
tc_labels
```

```python

centroids = kmeans.cluster_centers_
colors = ["blue", "red", "green", "yellow", "orange", "brown", "purple"]
for i in range(len(np.unique(kmeans.labels_))):
    plt.scatter(x = tc_labels[tc_labels["label"]==i].iloc[:,0], y = tc_labels[tc_labels["label"]==i].iloc[:,1] , color = colors[i])
plt.scatter(centroids[:,0] , centroids[:,1] , s = 80, color = "black")
```
